# 2 - C. Optimized Climbing Stairs Problem (with k steps)

Climbing Stairs Problem. There is a staircase and your goal is to answer the question, in how many distinct ways can you climb to the top. Assuming that we have k ways to climb a step i.e. 


•	One step at a time (-)

•	Two steps at one time (-)

•	k steps at one time (-)




```python
import sys
```

```python
class NegativeIntegerError(Exception):
    "Raised when the number passed to nsum function is a negative integer"
    pass

class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
```

```python
def GenerateBaseCases(k: int) -> list:
    try:
        if k < 0:
            raise NegativeIntegerError
        if k == 0:
            return [[1]]
        else:
            base_list = []
            for i in range(0,k+1):
                if len(base_list) >= 1:
                    prev_list = base_list[-1]
                    base_list.append([*prev_list[1:], sum(prev_list)])
                else:
                    base_list.append([*[0]*(k-1), 1])
            return  base_list
    except NegativeIntegerError:
        return("Exception Occured: Negative Integer provided, could not generate base cases")
```

```python
def steps(x: int, k: int, base_case: list, flag=True) -> list:
    if flag:
        flag = False
        if x > 100:
            sys.setrecursionlimit(x*30)
    try:
        if x < 0:
            raise NegativeIntegerError
        if (x >= 0) and (x <= len(base_case) - 1):
            return base_case[x]
        else:
            step_x1 = steps(x-1, k, base_case)
            return [*step_x1[-(k-1):], sum(step_x1)]
    
    except NegativeIntegerError:
        return("Exception Occured: Negative Integer found while calculating steps")
```

```python
def climbingSteps(x: int , k: int) -> str:
    calculated_cache_table = steps(x,k,GenerateBaseCases(k))
    sum_value = calculated_cache_table[-1] if type(calculated_cache_table) != type('str') else calculated_cache_table 
    return (f"The number of ways to climb {x} stairs is: {sum_value}")
```

```python
value = 50 # Enter the number of steps Value here
steps_permitted = 30 # Enter the permitted actions here
print(climbingSteps(value, steps_permitted))
```
