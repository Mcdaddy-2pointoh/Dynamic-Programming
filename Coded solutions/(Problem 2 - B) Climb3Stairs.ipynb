# 2 - B. Optimized Climbing Stairs Problem (with 3 steps)

Climbing Stairs Problem. There is a staircase and your goal is to answer the question, in how many distinct ways can you climb to the top. Assuming that we have three ways to climb a step i.e. 


•	One step at a time (-)

•	Two steps at one time (-)

•	Three steps at one time (-)




```python
import sys
```

```python
class NegativeIntegerError(Exception):
    "Raised when the number passed to nsum function is a negative integer"
    pass
```

```python
def steps(x: int, flag=True) -> list:
    if flag:
        flag = False
        if x > 100:
            sys.setrecursionlimit(x*30)
    try:
        if x < 0:
            raise NegativeIntegerError
        elif x == 0:
            return [0,0,1]
        elif x == 1:
            return [0,1,1]
        elif x == 2:
            return [1,1,2]
        elif x == 3:
            return [1,2,4]
        else:
            step_x1 = steps(x-1)
            return [*step_x1[-2:], sum(step_x1)]
    
    except NegativeIntegerError:
        return("Exception Occured: Negative Integer found")
```

```python
value = 10 # Enter your Value here
cache_table = steps(value)
sum_value = cache_table[-1] if type(cache_table) != type('str') else cache_table 
print(f"The number of ways to climb {value} stairs is: {sum_value}")
```
